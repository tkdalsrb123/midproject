```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd
import seaborn as sns



def pie_chart(country):
    labels = ['여가, 위락, 휴식','사업 또는 전문 활동','교육 (어학 프로그램, 연수 등)','종교 및 순례','기타']
    plt.rcParams['figure.figsize'] = (10, 10)
    plt.rcParams['font.size'] = 15
    plt.figure()
    plt.title(f'{country} 여행객들의 방문 목적 백분율')
    plt.pie(df_new_all.loc[f'{country}'], labels=labels, autopct='%.1f%%', labeldistance=1.2)
    plt.show();
```

```python
df_new_all = pd.read_csv('외래관광객 방문목적.csv').set_index('Unnamed: 0')
```

```python
pie_chart('전체')
```

방한 외국인들의 방문 목적을 나라별로 찾아볼 수 있는 파이차트

```python
df_a = df_new_all.drop('전체')
df_a
df_a = df_a.reset_index()
df_a.rename({'Unnamed: 0':'나라명'}, axis=1, inplace=True)
```

```python
df_melt = df_a.melt(id_vars='나라명',var_name='방문목적',value_name='방문자수')
df_melt
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df_melt, x=df_melt['나라명'], y=df_melt['방문자수'], hue='방문목적')
```

한눈에 확인해보는 나라별 방문 목적 그래프

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df_a, x=df_a['나라명'], y='교육 (어학 프로그램, 연수 등)')
```

교육을 들으러 오는 방한 외국인 수 비교. 전체 방문자수에 비례해서 대만,태국,말레이시아에서 많이들 온다.

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df_a, x=df_a['나라명'], y='종교 및 순례')
```
